```python
import requests

## api key 설정

KEY = '6f5a56764f646d7336327865795550'
SERVICE='energyUseDataSummaryInfo'

## 2015년 1월 ~ 2024년 12월 반복
for YEAR in range(2015, 2025):
    for MONTH in range(1, 13):
        ##데이터 수집 URL
         ##{MONTH:02d}=빈자리 수를 0으로 채우고 전체 수를 2자리로 맞춘 후 정수로 출력
        URL = f'http://openapi.seoul.go.kr:8088/{KEY}/json/energyUseDataSummaryInfo/1/5/{YEAR}/{MONTH:02d}'

       ##데이터 수집
        response = requests.get(URL)

        try:
            data = response.json()
            rows = data.get(SERVICE, {}).get('row', [])

            ## '개인' 데이터만 출력
            for item in rows:
                if item.get('MM_TYPE') == '개인':
                    ##현년 전기, 가스, 수도, 지역난방 에너지 사용량 출력
                    print(f"{YEAR}-{MONTH:02d}: "
                          f"전기={item.get('EUS', 'N/A')}, "
                          f"가스={item.get('GUS', 'N/A')}, "
                          f"수도={item.get('WUS', 'N/A')}, "
                          f"난방={item.get('HUS', 'N/A')}")

        ##실패출력
        except: print(YEAR, "- ", f"{MONTH:02d}", ": 데이터 수집 실패")
```
